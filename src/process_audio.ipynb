```python
import ffmpeg
import os

data_dir = "./unannotated/202508"
input_file = os.path.join(data_dir, "eval_1.MOV")
output_file = os.path.join(data_dir, "eval_1.mp4")

(
    ffmpeg
    .input(input_file)
    .output(output_file, vcodec="libx264", acodec="aac")
    .run()
)
```

```python
# to WAV
wav_output_file = os.path.join(data_dir, "eval_1.wav")
ffmpeg.input(input_file).output(wav_output_file, acodec="pcm_s16le", ar=44100).run()
```

```python
import subprocess

input_file = os.path.join(data_dir, "eval_1.MOV")
wav_output_file = os.path.join(data_dir, "eval_1_right.wav")

subprocess.run([
    "ffmpeg", "-i", input_file,
    "-map_channel", "0.1.1",  # 0=input file index, 1=audio stream index, 0=right channel
    wav_output_file
])
```
