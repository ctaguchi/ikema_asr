## Inter-annotator agreement

```python
import pympi

# eaf_file_YT = "unannotated/sinkoominkan_segmented_asr_1_YT.eaf"
eaf_file_YT = "unannotated/sinkoominkan_segmented_asr_1_YT2.eaf"
eaf_file_CT = "unannotated/sinkoominkan_segmented_asr_2.eaf"
eaf_file_ASR = "unannotated/sinkoominkan_segmented_asr.eaf"

eaf_YT = pympi.Elan.Eaf(eaf_file_YT)
annotations = eaf_YT.get_annotation_data_for_tier("default")
text_YT = [anno[2] for anno in annotations]
print(f"Number of annotations: {len(annotations)}")

eaf_CT = pympi.Elan.Eaf(eaf_file_CT)
annotations = eaf_CT.get_annotation_data_for_tier("default")
text_CT = [anno[2] for anno in annotations]
print(f"Number of annotations: {len(annotations)}")

eaf_ASR = pympi.Elan.Eaf(eaf_file_ASR)
annotations = eaf_ASR.get_annotation_data_for_tier("XLS-R-ELAN-Output")
text_ASR = [anno[2] for anno in annotations]
print(f"Number of annotations: {len(annotations)}")
```

```python
j = 0
k = 0
text_CT_clean = []
text_YT_clean = []
text_ASR_clean = []
for i in range(len(text_CT)):
    if text_CT[i] == "":
        j += 1
        continue
    
    if text_YT[j] != "":
        text_CT_clean.append(text_CT[i])
        text_YT_clean.append(text_YT[j].replace("　", " ").replace("  ", " "))
        text_ASR_clean.append(text_ASR[k])
    else:
        text_CT_clean.append(text_CT[i])
        text_YT_clean.append("x")
        text_ASR_clean.append(text_ASR[k])
    j += 1
    k += 1
    
print(f"Number of cleaned annotations: {len(text_CT_clean)}")
print(f"Number of cleaned YT annotations: {len(text_YT_clean)}")
print(f"Number of cleaned ASR annotations: {len(text_ASR_clean)}")
```

```python
import jiwer

wer = jiwer.wer(text_CT_clean, text_YT_clean)
wer_asr = jiwer.wer(text_CT_clean, text_ASR_clean)
wer_yt_asr = jiwer.wer(text_YT_clean, text_ASR_clean)

print(f"WER between YT and CT: {wer}")
print(f"WER between CT and ASR: {wer_asr}")
print(f"WER between YT and ASR: {wer_yt_asr}")
```

```python
cer = jiwer.cer(text_CT_clean, text_YT_clean)
cer_asr = jiwer.cer(text_CT_clean, text_ASR_clean)
cer_yt_asr = jiwer.cer(text_YT_clean, text_ASR_clean)

print(f"CER between YT and CT: {cer}")
print(f"CER between CT and ASR: {cer_asr}")
print(f"CER between YT and ASR: {cer_yt_asr}")
```
