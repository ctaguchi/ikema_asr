```python
import os
from pydub import AudioSegment
from tqdm import tqdm
import pympi
import datasets
import json

import utils

proj_dir = "/Users/chihirot/ikema_asr/"
data_dir = os.path.join(proj_dir, "data/")
src_dir = os.path.join(proj_dir, "src")
romaji_map_file = os.path.join(src_dir,
                               "romaji_map.json")
phoneme_map_file = os.path.join(src_dir,
                                 "phoneme_map.json")
with open(romaji_map_file, "r") as f:
    romaji_mapping = json.load(f)
with open(phoneme_map_file, "r") as f:
    phoneme_mapping = json.load(f)

prefix = "hougenkougi2_"
dir = "youtube/"

data = []
segment_dir = "tmp_segments/"
os.makedirs(
    os.path.join(
        data_dir,
        dir,
        segment_dir),
    exist_ok=True
)


with open(romaji_map_file, "r") as f:
    romaji_mapping = json.load(f)
with open(phoneme_map_file, "r") as f:
    phoneme_mapping = json.load(f)

for i in tqdm(range(1, 8)): # up to 7
    audio_prefix = f"{prefix}{i}"
    audio_path = os.path.join(data_dir,
                              dir,
                              audio_prefix,
                              audio_prefix + ".wav")
    eaf_path = os.path.join(data_dir,
                            dir,
                            audio_prefix,
                            audio_prefix + ".eaf")
    
    if os.path.exists(audio_path):
        sound = AudioSegment.from_wav(audio_path)
        eaf = pympi.Elan.Eaf(eaf_path)
        annotations = eaf.get_annotation_data_for_tier("default")
        if len(annotations[0]) == 4:
            annotations = [(start, end, text) for (start, end, text, _) in annotations]
            
        for idx, (start, end, text) in enumerate(annotations):
            transcription = utils.remove_tags(text)
            romaji = utils.hiragana_to_romaji(transcription, romaji_mapping)
            phoneme = utils.hiragana_to_phoneme(transcription, phoneme_mapping)

            if idx == 0:
                print(f"Example transcription: {transcription}")
                print(f"Example romaji: {romaji}")
                print(f"Example phoneme: {phoneme}")
            
            segment = sound[start:end]
            segment_file = f"{prefix}{i}_{start}_{end}.wav"
            segment_path = os.path.join(dir,
                                        segment_dir,
                                        segment_file)
            segment.export(segment_path, format="wav")
            print(f"Exported {segment_file} with text: {text}")
            data.append({"audio": segment_path,
                         "transcription": transcription,
                         "romaji": romaji,
                         "phoneme": phoneme,
                         "start": start,
                         "end": end,
                         "title": f"{prefix}{i}"})

        print(f"Processed {audio_path} with {len(annotations)} segments.")
    else:
        print(f"{audio_path} does not exist.")
        break
    
data[0]
```
